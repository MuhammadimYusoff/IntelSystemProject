# Data Preparation Section
Load the Iris data from SKLearn dataset example, reshape the data target and restructure the data to full dataframe

```python
# Import Dataset from SKLearn
from sklearn.datasets import load_iris

data = load_iris()
```

```python
import numpy as np

# Restructure the data value and reshape the data.target value to -1, 1 on axis 1
full_data = np.append(data.data, data.target.reshape(-1, 1), axis = 1)

# Printout the reshaped data
full_data
```

```python
import pandas as pd

df = pd.DataFrame(full_data)
df.columns = data.feature_names + ["type"]
```

```python
# Export the Iris dataframe to CSV file
df.to_csv("iris.csv")
```
